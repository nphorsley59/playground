```python
from math import e
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(z):
    return 1/(1+e**-z)
```

```python
x_sig = np.linspace(-5, 5, 100)
y_sig = sigmoid(x_sig)
```

```python
plt.plot(x_sig, y_sig)
plt.title("Sigmoid Activation Curve")
plt.xlabel("Neuron Output (z)")
plt.ylabel("Activation Value (a)")
plt.show()
```

```python
def tanh(z):
    return (e**z-e**-z)/(e**z+e**-z)
```

```python
x_tanh = np.linspace(-5, 5, 100)
y_tanh = tanh(x_tanh)
```

```python
plt.plot(x_tanh, y_tanh)
plt.title("Tanh Activation Curve")
plt.xlabel("Neuron Output (z)")
plt.ylabel("Activation Value (a)")
plt.show()
```

```python
def relu(z):
    return np.maximum(0, z)
```

```python
x_relu = np.linspace(-5, 5, 100)
y_relu = relu(x_relu)
```

```python
plt.plot(x_relu, y_relu)
plt.title("ReLU Activation Curve")
plt.xlabel("Neuron Output (z)")
plt.ylabel("Activation Value (a)")
plt.show()
```
